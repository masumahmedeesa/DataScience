```python
url = "https://en.wikipedia.org/wiki/List_of_Game_of_Thrones_episodes"
response = requests.get(url)
HTML(response.text)
soup = BeautifulSoup(response.text,'html5lib')
table = soup.find_all('table',{'class':'wikitable'})[-1]
table = table.find_all('tr')
table[0]
```

```python
scrap = []
for tb in table:
#     scrap.append([x.getText() for x in tb.find_all('th')])
    scrap.append([x.getText() for x in tb.find_all('td')])
del scrap[0:2]
scrap
```

```python
data = pd.DataFrame(scrap,columns=['E01','E02','E03','E04','E05','E06','E07','E08','E09','E10','Average'])
#data['Season'] = 1,2,3,4,5,6,7,8
col_value = [1,2,3,4,5,6,7,8]
data.insert(loc=0,column='Season',value=col_value)
data
```

```python
data.Average[6] = 10.26
data.Average[7] = 11.99
data.E10[6] = ""
data.E10[7] = ""
data.E09[6] = ""
data.E09[7] = ""
data.E08[6] = ""
data.E08[7] = ""
data.E07[7] = ""
data
```

```python
data.to_csv('Ratings.csv',index=False)
```
