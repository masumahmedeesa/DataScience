```python
from IPython.display import HTML
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
import numpy as np
import scipy as sp
import matplotlib as mp
from matplotlib import pyplot as plt
import matplotlib.cm as cm
import seaborn as sns
```

```python
url = "https://en.wikipedia.org/wiki/List_of_Game_of_Thrones_episodes"
response = requests.get(url)
HTML(response.text)
soup = BeautifulSoup(response.text, 'html5lib')
overView = soup.find_all('tr', {'class': 'vevent'})
overView
```

```python
tableTwoData = list()
for over in overView:
    tableTwoData.append([x.getText() for x in over.find_all('td')])
tableTwoData
```

```python
del tableTwoData[-1]
del tableTwoData[-1]
```

```python
df = pd.DataFrame(tableTwoData,
                  columns=['Episode_No_In_Season','Title','Directed_By', 'Written_By', 
                           'Original_Air_Date', 'US_Viewers_Million'])
```

```python
df
```

```python
df.US_Viewers_Million[0][:-4]
```

```python
df['US_Viewers_Million'] = df.US_Viewers_Million.apply(lambda x: x[:-4])
df.head()
```

```python
def last(ob):
    year = ob.split('\xa0')[-2]
    return year
last(df.Original_Air_Date[0])
df['Original_Air_Date_Year'] = df.Original_Air_Date.map(last)
df.head()
```

```python
df
```

```python
df.to_csv('GOT.csv', index=False)
```

```python
temp = data
temp.head()
```

```python
for director in directors:
    temp[director] = [director in d for d in temp.Directed_By]
temp.head()
```

```python
directors_work = temp[directors].sum()
directors_work
```

```python
directors_work.sort_values(ascending=False).plot(kind='bar')
```

```python
# for writters
writters = sorted(data.Written_By.unique())
# writters
for writter in writters:
    temp[writter] = [writter in w for w in temp.Written_By]
temp.head()

writters_written = temp[writters].sum()
writters_written
```

```python
writters_written.sort_values(ascending=False).plot(kind='bar')
```
